## Basics of Function Optimization

Let's consider a few functions. Find their optima (maxima and minima). Specify whether itâ€™s global optimum or local optimum, or both.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function definitions
def f1(x): return x**2
def f2(x): return x**3
def f3(x): return np.sin(x)
def f4(x): return x**4 - 4*x**2 + 4
def f5(x): return np.exp(x)
def f6(x): return x**2 - 4*x + 6 if x <= 2 else (x-3)**2 + 1

# X-axis ranges for each function
x1 = np.linspace(-2, 2, 1000)
x2 = np.linspace(-2, 2, 400)
x3 = np.linspace(-2*np.pi, 2*np.pi, 1000)
x4 = np.linspace(-3, 3, 1000)
x5 = np.linspace(-2, 1, 1000)
x6 = np.linspace(0, 5, 1000)

# Plotting
plt.figure(figsize=(12, 10))

plt.subplot(3, 2, 1)
plt.plot(x1, f1(x1))
plt.title("Quadratic Function (Global Minimum)")

plt.subplot(3, 2, 2)
plt.plot(x2, f2(x2))
plt.title("Cubic Function (No Global Minimum)")

plt.subplot(3, 2, 3)
plt.plot(x3, f3(x3))
plt.title("Trigonometric Function (Multiple Local Minima)")

plt.subplot(3, 2, 4)
plt.plot(x4, f4(x4))
plt.title("Polynomial Function (Local & Global Minima)")

plt.subplot(3, 2, 5)
plt.plot(x5, f5(x5))
plt.title("Exponential Function (Global Minimum)")

plt.subplot(3, 2, 6)
plt.plot(x6, [f6(xi) for xi in x6])
plt.title("Piecewise Function (Distinct Global & Local Minima)")

plt.tight_layout()
plt.show()
```

1. Quadratic Function (Global Minimum): This graph shows a single, global minimum at $\small x=0$.
2. Cubic Function (No Global Minimum): The cubic function's graph shows no global or local minima, illustrating how some functions don't have such points.
3. Trigonometric Function (Multiple Local Minima): The sine function demonstrates multiple local minima and global minima, at at $\small x=âˆ’Ï€,3Ï€,5Ï€,â€¦$.
4. Polynomial Function (Local & Global Minima): This graph depicts a function with both local and global minima at $\small x=Â±\sqrt2$ and a local maximum at $\small x=0$.
5. Exponential Function (Global Minimum): The exponential function graph approaches a global minimum as $x$ goes to negative infinity, but it never actually reaches a finite minimum value.
6. Piecewise Function (Distinct Global & Local Minima): This function shows a local minimum at $\small x=2$ and a global minimum at $\small x=3$, illustrating how different sections of a function can have distinct minima.
